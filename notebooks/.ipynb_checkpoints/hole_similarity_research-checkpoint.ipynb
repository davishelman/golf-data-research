# Hole Similarity Research

Explore the per-hole feature table and find similar golf holes across the 30 processed courses.

**Inputs** (build them first with `python -m pipeline.modeling all`):
- `courses/_index/hole_features.parquet` — one feature row per hole
- `courses/_index/hole_clusters.parquet` — cluster labels + PCA embedding
- `courses/_index/hole_similarity_examples.csv` — top-K nearest holes per hole

Requires: `pandas`, `matplotlib`, `scikit-learn` (`pip install pandas matplotlib scikit-learn`).

```python
from pathlib import Path
import pandas as pd

INDEX = Path('..') / 'courses' / '_index'
feat = pd.read_parquet(INDEX / 'hole_features.parquet')
clusters = pd.read_parquet(INDEX / 'hole_clusters.parquet')
examples = pd.read_csv(INDEX / 'hole_similarity_examples.csv')
print('holes:', len(feat), '| feature columns:', feat.shape[1])
feat.head()
```

## 1. Summary statistics

Coordinate frame: tee at `(0,0,0)`, green toward `+Y`, `z` is elevation relative to the tee. All `*_pct` columns are fractions in `[0,1]`.

```python
print('Courses:', feat['course_slug'].nunique(), '| Holes:', len(feat))
print('\nPar distribution:')
print(feat['par'].value_counts().sort_index())
cols = ['hole_length_m', 'fairway_pct', 'rough_pct', 'bunker_pct', 'water_pct',
        'trees_pct', 'dogleg_score', 'tee_to_green_elevation_change']
feat[cols].describe().round(3)
```

## 2. Clusters

KMeans and Agglomerative each produce 8 clusters over the scaled feature space. Below: cluster sizes and their golf character (par, length, surface mix).

```python
m = feat.merge(clusters[['hole_id', 'kmeans_cluster', 'agg_cluster', 'pca_1', 'pca_2']], on='hole_id')
summary = m.groupby('kmeans_cluster').agg(
    n=('hole_id', 'size'),
    mean_par=('par', 'mean'),
    mean_len_m=('hole_length_m', 'mean'),
    fairway=('fairway_pct', 'mean'),
    rough=('rough_pct', 'mean'),
    trees=('trees_pct', 'mean'),
    bunker=('bunker_pct', 'mean'),
    water=('water_pct', 'mean'),
    dogleg=('dogleg_score', 'mean'),
).round(3)
summary
```

## 3. PCA embedding

Each point is a hole, projected to the first two principal components and colored by KMeans cluster. (If you install `umap-learn` and rebuild, `umap_1/umap_2` columns will also be available.)

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(9, 7))
for cid, g in m.groupby('kmeans_cluster'):
    ax.scatter(g['pca_1'], g['pca_2'], s=14, alpha=0.8, label=f'cluster {cid}')
ax.set_xlabel('PC1'); ax.set_ylabel('PC2')
ax.set_title('Holes in PCA space (colored by KMeans cluster)')
ax.legend(loc='best', fontsize=8)
plt.tight_layout(); plt.show()
```

## 4. Query a hole → most similar holes

Look up the precomputed nearest neighbors for any `hole_id`.

```python
def show_similar(hole_id: str, n: int = 10) -> pd.DataFrame:
    rows = examples[examples['query_hole_id'] == hole_id].sort_values('rank').head(n)
    return rows[['rank', 'similar_hole_id', 'similar_course_slug',
                 'similar_hole_number', 'distance']].reset_index(drop=True)

show_similar('augusta_national__01', 10)
```

```python
# Compare the query hole with its top match side by side
compare_cols = ['hole_length_m', 'par', 'fairway_pct', 'rough_pct', 'bunker_pct',
                'water_pct', 'trees_pct', 'dogleg_score', 'fairway_width_drive_zone',
                'green_complex_bunker_pct', 'tee_to_green_elevation_change']
query = 'augusta_national__01'
match = show_similar(query, 1)['similar_hole_id'].iloc[0]
feat[feat['hole_id'].isin([query, match])].set_index('hole_id')[compare_cols].T
```

## 6. What the clusters mean (golf interpretation)

Read the `summary` table from section 2 alongside this guide:

- **Par-3 cluster** — low `mean_par` (~3) and short `mean_len_m` (~180–190 m). These holes have little/no drive zone, high green-complex hazard share.
- **Long par-4/par-5 cluster(s)** — high `mean_len_m`, substantial drive-zone fairway, larger elevation range.
- **Tree-lined vs open** — clusters split on `trees` percentage (parkland courses vs links/desert).
- **Water-heavy / penal** — elevated `water` and green-complex water; think island greens and forced carries.
- **Dogleg cluster** — high `dogleg` score and large `fairway_centerline_shift`.

Because every hole is in the same tee-relative aligned frame, two holes being close in feature space means they *play* similarly: comparable length, shot shape, hazard pressure (left vs right), and green-complex defense — regardless of which course they belong to.
